```python
def count_symbol(word):
    count = 0
    for i in word:
        if i in ['.', ',', '!', '?', '-','_','*','/','\\','@','#','$','%','^','&','(',')','+','=','{','}','[',']','|','~','`','<','>']:
            count += 1
    return count
```

```python
def nilai_label(x):
    if x>=90 and x <=100:
        return 'Buruk' 
    elif x>10 and x<90:
        return 'Sedang' 
    elif x>=0 and x <=10:
        return 'Baik' 
        
        
    
    
   
```
